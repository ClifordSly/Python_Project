```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

# load data
dataset=load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()

# data cleanup
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
```

```python
# lets create a new df we use copy to create a unique data frame

df_us = df[df['job_country']=='United States'].copy()
```

```python
df_us
```

```python
# To extract month from the table

df_us['job_posted_month']=df_us['job_posted_date'].dt.strftime('%B')

df_us
```

```python
df_us.pivot_table(
    index='job_posted_month',
    columns='job_title_short',
    aggfunc='size'
)
```

```python
# To arrange everything in chronoligical order
df_us_pivot=df_us.pivot_table(
    index='job_posted_month',
    columns='job_title_short',
    aggfunc='size'
)
```

```python
df_us_pivot.reset_index(inplace=True)
df_us_pivot['month_no'] = pd.to_datetime(
    df_us_pivot['job_posted_month'],
    format='%B'
)

df_us_pivot['month_no']
```

```python
df_us_pivot
```

```python
df_us_pivot.sort_values(by='month_no', inplace=True)
df_us_pivot.set_index('job_posted_month',inplace=True )
```

```python
df_us_pivot.drop(columns='month_no', inplace=True)
```

```python
df_us_pivot
```

```python
# lets plot

df_us_pivot.plot(kind='line')
```

```python
top_3=df_us['job_title_short'].value_counts().head(3)
top_3 =top_3.index.tolist()
top_3
```

```python
df_us_pivot[top_3].plot(kind='line')
plt.title('Monthly Job Postings for Top Data Jobs in the US')
plt.xlabel('2023')
plt.ylabel('job Count')
plt.legend
plt.show()
```
