```python
# importing libraries

import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#loading data
dataset=load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()

#data cleanup $ convert job skills column from a string to a list
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
df['job_skills']=df['job_skills'].apply(lambda x:ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df['job_work_from_home'].value_counts().plot(kind='pie', startangle=180, autopct='%1.1f%%')
plt.title('Work From Home Status')
plt.ylabel('')
plt.show()
```

### I should always be carefull with the type of visual I use

```python
df['job_title_short'].value_counts().plot(kind='pie')
plt.title('Job Titles in Cliford Dataset')
plt.ylabel('')
plt.show()
```

```python
df['job_title_short'].value_counts().plot(kind='barh')
plt.title('Job Titles in Cliford Dataset')
plt.ylabel('')
plt.gca().invert_yaxis()
plt.show()
```

### Matplotlib pie scatter plot

```python
df[['job_work_from_home','job_no_degree_mention', 'job_health_insurance']]
```

```python
fig, ax=plt.subplots(1,3)

dict_column={
  'job_work_from_home':'Work from Home',
  'job_no_degree_mention':'Job Degree Req.',
  'job_health_insurance':'Health Insurance Offred' 
}

for i, (column, title) in enumerate(dict_column.items()):
  ax[i].pie(df[column].value_counts(), startangle=180, autopct='%1.1f%%', labels=['False','True'])
  ax[i].set_title(title)

plt.show()
```
