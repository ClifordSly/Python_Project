```python
# install datasets library (if not already installed)
!pip install datasets

# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt


# Loading Data
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
x= [1, 2, 3, 4]
y= [1, 2, 3, 4]

plt.plot (x,y)
plt.show()
```

# Plotting for job_posted_date

```python
df.job_posted_date.head()
```

```python
plt.plot(df.job_posted_date, df.job_posted_date)
plt.show()
```

```python
# let's do some aggregation then assign a data frame

date_counts = df.job_posted_date.value_counts()

date_counts.sort_index(inplace=True)
```

```python
date_counts.head()
```

```python
plt.plot(date_counts.index, date_counts)
plt.show()
#
```

```python
# above aggregate  was not correct we need to aggregate  by month, day or year. Lets use month and create job_posted_month column

df['job_posted_month'] = df['job_posted_date'].dt.month
```

```python
monthly_counts = df.job_posted_month.value_counts()

monthly_counts.sort_index(inplace=True)

monthly_counts
```

```python
plt.plot(monthly_counts.index, monthly_counts)
plt.show()
#
```

# Bar Chart- Counts of Job Postings

```python
df.job_title_short.value_counts()

job_counts = df.job_title_short.value_counts().head(10)

job_counts
```

```python

plt.bar(job_counts.index, job_counts)
plt.title('Postings by Job Title')
plt.xlabel('Job Title')
plt.ylabel('Number of Postings')
plt.xticks(rotation=45, ha= 'right')
plt.show()
```

For horizontal bar plot

```python
job_counts= job_counts.sort_values(ascending=True)

plt.barh(job_counts.index, job_counts)
plt.show()
#
```
